**INFORMATION**

The skewness of a data distribution is a statistical measure that helps understand whether a distribution is symmetrical and how it is shaped around the mean. Skewness is defined by the long tails of the distribution leaning to one side and can be positive or negative.

In a positive (**Right Skewed**) distribution, most of the data is concentrated in the low values, but a few high values ​​lie towards the right tail of the distribution.

In a negative (**Left Skewed**) distribution, most of the data is concentrated in the high values, but a few low values ​​lie towards the left tail of the distribution.

In a symmetric (**Zero Skew**) distribution, the distribution is symmetric, meaning the data is evenly distributed around the mean. The normal distribution is such a distribution and the skewness value is close to zero.

In other words;

Skewness = 0: Perfectly symmetric distribution (normal distribution)
Skewness > 0: Positive skewed (right skewed) distribution
Skewness < 0: Negative skewed (left skewed) distribution

The choice of median or mean strategy for filling missing values ​​depends on the skewness of the distribution of the data set. Since skewness indicates whether the data set is symmetric or not, it plays an important role in deciding which filling strategy is more appropriate.

When the skewness value is between 0 and ±0.5, the mean strategy may be appropriate, since in this range the distribution is usually symmetric or very slightly skewed.

If the skewness value is greater than |0.5| (i.e. less than -0.5 or greater than 0.5), padding with the median is usually a more accurate strategy.

```python
import matplotlib.gridspec as gridspec
from scipy import stats
```

```python
def three_chart_plot(df, feature):
    fig = plt.figure(constrained_layout = True, figsize = (12, 8))
    grid = gridspec.GridSpec(ncols = 3, nrows = 3, figure = fig)

    ax1 = fig.add_subplot(grid[0, :2])
    ax1.set_title('Histogram Graph')

    sns.distplot(df.loc[:, feature], norm_hist = True, ax = ax1)
    plt.axvline(x = df[feature].mean(), c = 'red')
    plt.axvline(x = df[feature].median(), c = 'green')

    ax2 = fig.add_subplot(grid[1, :2])
    ax2.set_title('QQ_plot')
    stats.probplot(df.loc[:,feature], plot = ax2)

    ## Customizing the Box Plot.
    ax3 = fig.add_subplot(grid[:, 2])
    ## Set title.
    ax3.set_title('Box Plot')
    sns.boxplot(df.loc[:,feature], orient = 'v', ax = ax3 )
```

```python
from sklearn.model_selection import cross_validate
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
y = df_train['SalePrice']
X = df_train.drop(['SalePrice'], axis=1)
```

```python
X = pd.get_dummies(X)
df_test_encoded = pd.get_dummies(df_test)
```

```python
X, df_test_encoded = X.align(df_test_encoded, join='left', axis=1, fill_value=0)
```

```python
cart_model = DecisionTreeRegressor(random_state=5)
```

```python
cv_results = cross_validate(cart_model, X, y, cv=5, scoring=["r2", "neg_mean_squared_error", "neg_mean_absolute_error"])

print("R-squared score: ", cv_results['test_r2'].mean())
print("Mean squared error: ", -cv_results['test_neg_mean_squared_error'].mean())
print("Mean absolute error: ", -cv_results['test_neg_mean_absolute_error'].mean())
```

```python
print("AVG Price for train:", df_train['SalePrice'].mean())
```

As you can see, our R² score was calculated as 0.7189. The R² score shows how well the model fits the data. It shows that the model explains approximately 71.89% of the variance between the independent variables and the target variable. The MAE value was calculated as 26.466. The MAE shows the average deviation in the model's predictions. When we look at the average of the SalePrice target variable in 'df_train', it was calculated as 180.921.

Now let's try to increase the success of our model by doing hyperparameter optimization and build the final model with the best parameters. We will not do detailed hyperparameter optimization here.

```python
from sklearn.model_selection import GridSearchCV

param_grid = {
    'max_depth': [None, 5, 10, 15, 20],
    'min_samples_split': [2, 5, 10],
    'max_features': [None, 'auto', 'sqrt', 'log2'],
    'min_samples_leaf': [1, 2, 5, 10]
}

grid_search = GridSearchCV(estimator=cart_model,
                           param_grid=param_grid,
                           cv=5,
                           scoring='neg_mean_squared_error',
                           n_jobs=-1,
                           verbose=1)

grid_search.fit(X, y)
```

```python
print("Best Parameters: ", grid_search.best_params_)
```

```python
cart_final = grid_search.best_estimator_
```

```python
cv_results = cross_validate(cart_final, X, y, cv=5, scoring=["r2", "neg_mean_squared_error", "neg_mean_absolute_error"])

print("R-squared score: ", cv_results['test_r2'].mean())
print("Mean squared error: ", -cv_results['test_neg_mean_squared_error'].mean())
print("Mean absolute error: ", -cv_results['test_neg_mean_absolute_error'].mean())
```

The new scores we obtained after hyperparameter optimization clearly show the improvement in the model. The increase in the R² score shows that the model has improved. The model now explains 77.33% of the target variable. We can observe that the MSE value has also decreased. There is also an improvement in the MAE value. The average error in the predictions has decreased from 26,466.09 to 24,359.77.

Let's look at the variable importance scores provided by the model and examine the most important variables by visualizing them.

```python
def plot_importance(model, features, num=len(X)):
    feature_imp = pd.DataFrame({'Value': model.feature_importances_, 'Feature': features.columns})
    plt.figure(figsize=(10, 10))
    sns.set(font_scale=1)
    sns.barplot(x="Value", y="Feature", data=feature_imp.sort_values(by="Value", ascending=False)[0:num])

    plt.title('Features')
    plt.tight_layout()
    plt.show()
```

```python
plot_importance(cart_final, X, num=10)
```
